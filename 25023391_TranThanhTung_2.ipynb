Trường hợp 1:

1. Nguyên nhân lỗi
Trong code có dòng: result.append(item / 0)

→ gây ZeroDivisionError vì chia cho 0.



2. Giải pháp fix

Không nên chia cho 0.

Nếu muốn xử lý số âm thì có thể:

Ghi lại là None, hoặc

Gán giá trị mặc định, hoặc

Ném ra lỗi có thông báo rõ ràng.

3. Code fix với error handling

```python
def process_data(data):
    result = []
    for item in data:
        try:
            if item > 0:
                result.append(item * 2)
            else:
                # thay vì chia 0, ta đưa giá trị None
                result.append(None)
        except Exception as e:
            print(f"Lỗi khi xử lý {item}: {e}")
            result.append(None)
    return result

# Test
data = [1, -2, 3, 0, -5]
print(process_data(data))
```

4. Test cases

```python
print(process_data([1, 2, 3]))        # [2, 4, 6]
print(process_data([-1, -2]))         # [None, None]
print(process_data([0]))              # [None]
print(process_data([]))               # []
```

3. Code fix với error handling

```python
def find_max_number(numbers):
    if not numbers:
        raise ValueError("Danh sách rỗng, không thể tìm max")
    max_num = numbers[0]
    for num in numbers:
        if num > max_num:
            max_num = num
    return max_num

# Test
numbers = [-5, -10, -2, -1]
print(find_max_number(numbers))  # -1
```

4. Test cases

```python
print(find_max_number([1, 2, 3]))        # 3
print(find_max_number([-5, -10, -2]))    # -2
print(find_max_number([0]))              # 0
try:
    print(find_max_number([]))           # ValueError
except ValueError as e:
    print("Caught:", e)
```

3. Code fix với error handling

```python
def process_student_scores(students, subject_index):
    results = []
    for student in students:
        try:
            name = student[0]
            scores = student[1]

            # lọc điểm hợp lệ (int, float)
            numeric_scores = [s for s in scores if isinstance(s, (int, float))]
            if not numeric_scores:
                raise ValueError(f"Sinh viên {name} không có điểm hợp lệ")

            # kiểm tra subject index
            if subject_index >= len(scores):
                subject_score = None
            else:
                subject_score = scores[subject_index] if isinstance(scores[subject_index], (int, float)) else None

            # tính trung bình
            average = sum(numeric_scores) / len(numeric_scores)
            grade = calculate_grade(average)

            results.append({
                'name': name,
                'subject_score': subject_score,
                'average': average,
                'grade': grade
            })
        except Exception as e:
            print(f"Lỗi khi xử lý sinh viên {student}: {e}")
            results.append({'name': student[0], 'error': str(e)})
    return results

def calculate_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

# Test data
students = [
    ['Alice', [85, 90, 78]],
    ['Bob', [92, 88]],              # ít môn
    ['Charlie', ['N/A', 85, 90]],   # có string
    ['David', []]                   # rỗng
]
print(process_student_scores(students, 2))
```

4. Test cases

```python
print(process_student_scores([['Eve', [100, 95, 90]]], 1))   # subject_score = 95
print(process_student_scores([['Tom', [50, 40, 30]]], 0))    # grade = F
print(process_student_scores([['Jerry', []]], 0))            # error case
print(process_student_scores([['Sam', ['N/A', 70]]], 1))     # xử lý 'N/A'
```
