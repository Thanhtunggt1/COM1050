```python
#BAI1
toan = float(input('Diem mon Toan'))
ly = float(input('Diem mon Ly'))
hoa = float(input('Diem mon Hoa'))
anh = float(input('Diem mon Anh'))
diemtrungbinh = (toan+ly+hoa+anh)/4 
if 10>=diemtrungbinh>=9:
    print('Xuat sac')
elif 9>diemtrungbinh>=8.5:
    print('Gioi')
elif 8.5>diemtrungbinh>=7:
    print('Kha')
elif 7>diemtrungbinh>=5.5:
    print('Trung binh')
elif 5.5>diemtrungbinh>=4:
    print('Yeu')
else:
    print('Kem')
```

```python
#BAI 2 
cao = float(input('Chieu cao'))
nang = float(input('Can nang'))
bmi = nang/(cao*cao)
if 0<bmi<18.5:
    print('Thieu can')
elif 25>=bmi>=18.5:
    print('Binh thuong')
elif 30>=bmi>=25:
    print('Thua can')
elif bmi>=30:
    print('Thua can')
print('BMI cua ban la: %.2f' %bmi)
```
