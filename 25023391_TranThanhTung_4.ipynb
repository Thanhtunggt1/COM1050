**Bài 4: ** Viết chương trình tính phí vận chuyển:
- Nhập trọng lượng (kg)
- Nhập khoảng cách (km)
- Nhập có giao nhanh không (y/n)

Phí cơ bản:
- Dưới 1kg: 20,000đ
- 1-5kg: 40,000đ
- Trên 5kg: 60,000đ

Phụ phí khoảng cách (nếu > 10km):
- Mỗi km thêm: 2,000đ

Phí giao nhanh (nếu chọn):
- Cộng thêm 50%

In ra chi tiết:
- Phí cơ bản: ...
- Phụ phí khoảng cách: ...
- Phí giao nhanh: ...
- Tổng cộng: ...

```python
weight = float(input("Nhập trọng lượng (kg): "))
distance = float(input("Nhập khoảng cách (km): "))
express = input("Giao nhanh? (y/n): ")
# Viết code ở đây
if weight < 1:
    phicoban = 20000
elif weight <= 5:
    phicoban = 40000
else:
    phicoban = 60000

if distance > 10:
    phikhoangcach = (distance - 10) * 2000
else:
    phikhoangcach = 0

if express == 'y':
    phivanchuyen = 0.5 * (phicoban + phikhoangcach)
elif express == 'n':
    phivanchuyen = 0

tong = phicoban + phikhoangcach + phivanchuyen
print(f'Phí cơ bản: {phicoban:.0f}đ')
print(f'Phụ phí khoảng cách: {phikhoangcach:.0f}đ')
print(f'Phí giao nhanh: {phivanchuyen:.0f}đ')
print(f'Tổng cộng: {tong:.0f}đ')
```

- Nhập tổng chi tiêu (VND)
- Nhập số năm thành viên
- Nhập số lần khiếu nại

Quy tắc:
1. Nếu khiếu nại >= 5: "Đồng"
2. Nếu khiếu nại < 5:
   - Chi tiêu >= 50tr:
     * Năm >= 3: "Bạch Kim"
     * Năm < 3: "Vàng"
   - 20tr <= Chi tiêu < 50tr:
     * Năm >= 2: "Vàng"
     * Năm < 2: "Bạc"
   - Chi tiêu < 20tr:
     * Năm >= 1: "Bạc"
     * Năm < 1: "Đồng"

Test với nhiều trường hợp khác nhau

```python

total_spent = float(input("Nhập tổng chi tiêu: "))
years = int(input("Nhập số năm thành viên: "))
complaints = int(input("Nhập số lần khiếu nại: "))
# Viết code ở đây
if complaints >= 5:
    rank = 'Đồng'
else:
    if total_spent >= 50:
        if years >= 3:
            rank = 'Bạch Kim'
        else:
            rank = "Vàng"
    elif 20 <= total_spent < 50:
        if years >= 2:
            rank = 'Vàng'
        else:
            rank = 'Bạc'
    else:
        if years >= 1:
            rank = 'Bạc'
        else:
            rank = 'Đồng'
print(f'Hạng của khách hàng là: {rank}')
```

**Bài 6:** Viết chương trình tính tiền điện theo bậc thang:
- Bậc 1: 0-50 kWh: 1,678đ/kWh
- Bậc 2: 51-100 kWh: 1,734đ/kWh
- Bậc 3: 101-200 kWh: 2,014đ/kWh
- Bậc 4: 201-300 kWh: 2,536đ/kWh
- Bậc 5: 301-400 kWh: 2,834đ/kWh
- Bậc 6: Trên 400 kWh: 2,927đ/kWh

Input:
- Chỉ số cũ
- Chỉ số mới

Output:
- Điện tiêu thụ: ... kWh
- Tiền bậc 1: ...
- Tiền bậc 2: ...
- ...
- Tổng tiền (chưa VAT): ...
- VAT (10%): ...
- Tổng thanh toán: ...

```python
old_index = int(input("Chỉ số cũ: "))
new_index = int(input("Chỉ số mới: "))
#Viết code owe đây
u = new_index - old_index
bac1 = 1678
bac2 = 1734
bac3 = 2014
bac4 = 2536
bac5 = 2834
bac6 = 2927
if u <= 50:
    b1 = u * bac1
    b2 = b3 = b4 = b5 = b6 = 0
elif u <= 100:
    b1 = 50 * bac1
    b2 = (u - 50) * bac2
    b3 = b4 = b5 = b6 = 0
elif u <= 200:
    b1 = 50 * bac1
    b2 = 50 * bac2
    b3 = (u - 100) * bac3
    b4 = b5 = b6 = 0
elif u <= 300:
    b1 = 50 * bac1
    b2 = 50 * bac2
    b3 = 100 * bac3
    b4 = (u - 200) * bac4
    b5 = b6 = 0
elif u <= 400:
    b1 = 50 * bac1
    b2 = 50 * bac2
    b3 = 100 * bac3
    b4 = 100 * bac4
    b5 = (u - 300) * bac5
    b6 = 0
else:
    b1 = 50 * bac1
    b2 = 50 * bac2
    b3 = 100 * bac3
    b4 = 100 * bac4
    b5 = 100 * bac5
    b6 = (u - 400) * bac6
tong = b1 + b2 + b3 + b4 + b5 + b6
vat = tong * 0.10
Ketqua = tong + vat
print(f'Điện tiêu thụ: {u} kWh')
print(f'Tiền bậc 1: {b1:} đ')
print(f'Tiền bậc 2: {b2:} đ')
print(f'Tiền bậc 3: {b3:} đ')
print(f'Tiền bậc 4: {b4:} đ')
print(f'Tiền bậc 5: {b5:} đ')
print(f'Tiền bậc 6: {b6:} đ')
print(f'Tổng tiền (chưa VAT): {tong:} đ')
print(f'VAT (10%): {vat:.0f} đ')
print(f'Tổng thanh toán: {Ketqua:.0f} đ')
```

**Bài 8:**  Viết chương trình tính lương:

Input:
- Số giờ làm việc
- Lương theo giờ
- Loại ngày (weekday/weekend/holiday)
- Đánh giá (poor/average/good/excellent)

Quy tắc:
1. Tính lương theo loại ngày:
   - Weekday:
     * 8 giờ đầu: x1.0
     * Giờ 9-10: x1.5
     * Từ giờ 11: x2.0
   - Weekend: toàn bộ x1.5
   - Holiday: toàn bộ x2.0

2. Thưởng/phạt hiệu suất:
   - poor: -10%
   - average: 0%
   - good: +10%
   - excellent: +20%

3. Thưởng chuyên cần:
   - Làm >= 48 giờ: +500,000đ

Output chi tiết:
- Lương giờ hành chính: ...
- Lương tăng ca: ...
- Tổng lương cơ bản: ...
- Điều chỉnh hiệu suất: ...
- Thưởng chuyên cần: ...
- Tổng lương: ...

```python

hours = float(input("Số giờ làm việc: "))
hourly_rate = float(input("Lương theo giờ: "))
day_type = input("Loại ngày (weekday/weekend/holiday): ")
performance = input("Hiệu suất (poor, average, good, excellent): ")
#Viết code ở đây

if day_type == 'weekday':
    if hours <= 8:
        a = 0
    elif 11 >= hours:
        a = 0.5 * (hours - 8) * hourly_rate
    elif hourly_rate > 11:
        a = 3 * 0.5 * hourly_rate + (hours - 11) * 1 * hourly_rate
elif day_type == 'weekend':
    a = 0.5 * hourly_rate * hours
elif day_type == 'holiday':
    a = 1 * hourly_rate * hours

if performance == 'poor':
    b = -0.1
elif performance == ' average':
    b = 0
elif performance == 'good':
    b = 0.1
elif performance == 'excellent':
    b = 0.2

if hours >= 48:
    c = 500000
else:
    c = 0

print(f'Lương giờ hành chính: {hours * hourly_rate:.0f}đ')
print(f'Lương tăng ca: {a:.0f}')
print(f'Tổng lương cơ bản: {hours * hourly_rate + a:.0f}đ')
print(f'\nĐiều chỉnh hiệu suất: {b * 100:.0f}%')
print(f'Thưởng chuyên cần: {c:.0f}đ')
print(f'Tổng lương: {(hourly_rate * hours + a + c) * (1 + b):.0f}đ')
```

Bài 9: Chương trình sau có lỗi. Hãy:
1. Thêm các dòng print để trace control flow
2. Tìm và sửa lỗi
3. Giải thích lỗi là gì?

```python
price = float(input("Giá sản phẩm: "))
quantity = int(input("Số lượng: "))
customer_type = input("Loại khách (VIP/Regular): ")

discount = 0

if customer_type == "VIP":
    if quantity > 10:
        discount = 0.2
    else:
        discount = 0.1
elif customer_type == "Regular":
    if quantity > 20:
        discount = 0.15
    else:
        discount = 0.05

# Lỗi ở đây!  
final_price = price * quantity * discount
print(f"Giá cuối: {final_price}")
```

TÌM VÀ SỬA LỖI 

Lỗi ở dòng: final_price = price * quantity * discount
Sửa thành: final_price = price * quantity * (1 - discount)

GIẢI THÍCH LỖI

Dòng: final_price = price * quantity * discount tính số tiền được giảm, không phải giá sau giảm

```python
#THÊM CÁC DÒNG PRINT ĐỂ TRACE CONTROL FLOW

price = float(input("Giá sản phẩm: "))
quantity = int(input("Số lượng: "))
customer_type = input("Loại khách (VIP/Regular): ")

discount = 0
print(f'Bắt đầu tính giá cho khách {customer_type}')
print(f'Giá của sản phẩm: {price:.0f}đ')
print(f'Số lượng: {quantity}')

if customer_type == 'VIP':
    print('Loại khách: VIP')
    if quantity > 10:
        discount = 0.2
        print('Mua trên 10 sản phẩm, giảm 20%')
    else:
        discount = 0.1
        print('Mua ≤ 10 sản phẩm, giảm 10%')

elif customer_type == 'Regular':
    print('Loại khách: Regular')
    if quantity > 20:
        discount = 0.15
        print('Mua trên 20 sản phẩm, giảm 15%')
    else:
        discount = 0.05
        print('Mua ≤ 20 sản phẩm, giảm 5%')

else:
    print('Loại khách không hợp lệ, không có giảm giá')


final_price = price * quantity * (1 - discount)

print(f'Tổng tiền trước giảm: {price * quantity:.0f}đ')
print(f'Tỉ lệ giảm giá: {discount*100:.0f}%')
print(f'Giá cuối cùng: {final_price:.0f}đ')
```
