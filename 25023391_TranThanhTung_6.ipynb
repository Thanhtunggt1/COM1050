**Bài 1:** Viết hàm triangle_type(a, b, c) nhận vào 3 cạnh và trả về loại tam giác:



*   "Không phải tam giác" nếu không thỏa điều kiện tam giác
*  "Tam giác đều" nếu cả 3 cạnh bằng nhau
*  "Tam giác cân" nếu có 2 cạnh bằng nhau
*   "Tam giác vuông" nếu thỏa định lý Pythagoras
*   "Tam giác thường" trong các trường hợp còn lại






Yêu cầu:

Sử dụng nested if statements
Viết test cases cho mỗi loại tam giác

```python
def triangle_type(a, b, c):
    # Code của bạn ở đây
    if a + b >= c and b + c >= a and c + a >= b:
        if a == b and b == c:
            return "Tam giac deu"
        else:
            if a == b or a == c or c == b:
                return "Tam giac can"
            else :
                if a ** 2 + b ** 2 == c ** 2 or b ** 2 + c ** 2 == a ** 2 or a ** 2 + c ** 2 == b ** 2:
                   return "Tam giac vuong"
                else:
                    return "Tam giac thuong"           
    else :
        return "Khong phai tam giac" 

print(triangle_type(3, 4, 5))
```

**Bài 2:** Viết hàm calculate_shipping(weight, distance, is_express) tính phí vận chuyển:
Quy tắc:

Phí cơ bản:



*   Nếu weight <= 1kg: 20,000 đồng
*   Nếu 1kg < weight <= 5kg: 40,000 đồng

*   Nếu weight > 5kg: 60,000 đồng

Phụ phí khoảng cách:

*   Nếu distance > 100km: thêm 10,000 đồng
*   Nếu distance > 500km: thêm 30,000 đồng


Phụ phí express:

Nếu is_express = True: nhân hệ số 1.5



Yêu cầu:

Viết hàm trả về tổng phí vận chuyển
Viết ít nhất 8 test cases để cover các trường hợp

```python
def calculate_shipping(weight, distance, is_express):
    # Code của bạn ở đây
    if 0 < weight <= 1:
        a = 20000
    elif weight <= 5:
        a = 40000
    else :
        a = 60000
    
    if  distance  > 500:
        b = 30000
    elif distance > 100:
        b = 10000
    else : 
        b = 0
    
    if is_express == True:
        c = 1.5
    else :
        c = 1 

    return(f"{(a + b) * c:,.0f} VND")           
# Test cases
# Case 1: weight=0.5, distance=50, is_express=False -> Expected: 20000
# Case 2: weight=3, distance=150, is_express=False -> Expected: 50000
# Thêm các test cases khác...
print(calculate_shipping(0.5, 50, False))
print(calculate_shipping(1, 200, False))
print(calculate_shipping(0.5, 600, False))
print(calculate_shipping(3, 80, False))
print(calculate_shipping(3, 200, False))
print(calculate_shipping(10, 800, False))
print(calculate_shipping(4, 50, True))
print(calculate_shipping(7, 1000, True))
```

**Bài 3:** Viết hàm is_valid_date(day, month, year) kiểm tra ngày có hợp lệ không.
Quy tắc:

Năm nhuận: chia hết cho 4 NHƯNG không chia hết cho 100, HOẶC chia hết cho 400.
Tháng 1,3,5,7,8,10,12: có 31 ngày.
Tháng 4,6,9,11: có 30 ngày.
Tháng 2: 29 ngày nếu năm nhuận, 28 ngày nếu không.

Yêu cầu:

Trả về True/False
Viết test cases cho: năm nhuận, năm không nhuận, các tháng khác nhau, các trường hợp biên

```python
def is_valid_date(day, month, year):
    # Code của bạn ở đây
    if year < 1:
        return False
    
    if month in [1, 3, 5, 7, 8, 10, 12]:
        a = 31
    elif month in [4, 6, 9, 11]:
        a = 30
    elif month == 2:
        if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
            a = 29
        else:
            a = 28
    else:
        return False

    return 0 < day <= a

# Test cases
print(is_valid_date(29, 2, 2024))  # True (năm nhuận)
print(is_valid_date(29, 2, 2023))  # False (không nhuận)
print(is_valid_date(31, 4, 2024))  # False (tháng 4 chỉ 30 ngày)
# Thêm các test cases khác...
print(is_valid_date(31, 12, 0)) 
print(is_valid_date(0, 1, 2024))   
print(is_valid_date(15, 13, 2022)) 
print(is_valid_date(28, 2, 1900))  
print(is_valid_date(29, 2, 2000))   
```

**Bài 4:** Viết hàm suggest_outfit(temperature, is_raining, is_windy, event_type) gợi ý trang phục.
Input:


*   temperature: nhiệt độ (°C)

*   is_raining: True/False
*   is_windy: True/False


*   event_type: "formal", "casual", "sport"


Quy tắc:

Xét event_type trước.
Với mỗi loại event, xét thời tiết:


*   Nếu temperature < 15: đề xuất mặc ấm

*   Nếu is_raining: đề xuất áo mưa/ô
*   Nếu is_windy: đề xuất áo khoác




Yêu cầu:

Sử dụng nested if statements
Trả về string mô tả trang phục
Viết 12 test cases để cover tất cả các branch

```python
def suggest_outfit(temperature, is_raining, is_windy, event_type):
    # Code của bạn ở đây
    outfit = ""
    if event_type == "formal":
        outfit = "Bo vest thanh lich"
        if temperature < 15:
            outfit += ", mac them ao am"
        if is_raining:
            outfit += ", mang o hoac ao mua"
        if is_windy:
            outfit += ", khoac them ao ngoai"

    elif event_type == "casual":
        outfit = "Trang phuc thoai mai"
        if temperature < 15:
            outfit += ", mac them ao khoac am"
        if is_raining:
            outfit += ", mang ao mua nhe"
        if is_windy:
            outfit += ", khoac ao gio"

    elif event_type == "sport":
        outfit = "Do the thao nang dong"
        if temperature < 15:
            outfit += ", mac ao giu nhiet"
        if is_raining:
            outfit += ", chon do chong nuoc"
        if is_windy:
            outfit += ", mac ao khoac nhe"

    else:
        outfit = "Loai su kien khong hop le"

    return outfit
#Test case
print(suggest_outfit(10, False, False, "formal"))  
print(suggest_outfit(25, True, False, "formal"))
print(suggest_outfit(18, False, True, "formal"))   
print(suggest_outfit(8, True, True, "formal"))      

print(suggest_outfit(14, False, False, "casual")) 
print(suggest_outfit(22, True, False, "casual"))
print(suggest_outfit(20, False, True, "casual")) 
print(suggest_outfit(5, True, True, "casual"))    

print(suggest_outfit(13, False, False, "sport"))    
print(suggest_outfit(28, True, False, "sport"))    
print(suggest_outfit(19, False, True, "sport")) 
print(suggest_outfit(12, True, True, "sport")) 
```
