```python
#W6A1
a = list(map(int, input().split()))
print(a)
b=[]
for i in range(len(a)):
    if a[i] not in b:
        b.append(a[i])

print(b)
```

```python
#W6A5
a = tuple(map(int, input().split()))
x = y = z = 0
print(a)

def dem_so(a):
    global x, y, z
    for i in a:
        if i > 0:
            x += 1
        elif i < 0:
            y += 1
        else:
            z += 1

dem_so(a)
print({'positives': x, 'negatives': y, 'zeros': z})
```
